#### üëâ As we know the team that wins the toss they gets to choose whether they want to bowl first or bat first, they chooses this based on the pitch conditions. Lets compare the chances of winning a game by the toss winning team.

```python
winning_team = df1.toss_winner == df1.winner
```

```python
win = winning_team.value_counts()
```

```python
win
```

```python
df1.id.count()
```

```python
win1 = win.index
```

#### üëâ Total matches that were played are 756 out of which 393 matches were won by the toss winning team and 363 matches were won by the team that lost the toss.

#### üëâ converting into percentage factor

```python
toss_winning_win_percentage = (393/756)*100
toss_losing_win_percentage = (363/756)*100
print(f"Toss winning win percentage is {round(toss_winning_win_percentage,3)}%")
print(f"Toss losing win percentage is {round(toss_losing_win_percentage,3)}%")
```

#### üëâ We can observe that the factor of winning a toss does not put much effect to the win rate of the team as the team winning toss has only a win rate percentage difference of 3% approx.

#### üëâ plotting pie chart

```python
colors = ['gold', 'lightcoral']
plt.pie(win, labels = win1, autopct='%1.1f%%', shadow=True, startangle=140, colors = colors)
```

### üëâ Maximum wickets taken by bowlers

```python
df2.sample(5)
```

```python
df2.info()
```

```python
df2['dismissal_kind'].unique()
```

#### üëâ Run out and retired hurt doesn't counts in favour of the bowler. So these types are eliminated

```python
dismissed = df2[df2.dismissal_kind != 'run out']
dismissed1 = dismissed[dismissed.dismissal_kind != 'retired hurt']
```

```python
dismissed1.info()
```

```python
dismissed1.sample(5)
```

```python
total_dismissed = dismissed1.groupby('bowler').apply(lambda x : x['dismissal_kind'].dropna()).reset_index(name='Wickets')
```

```python
count_of_wickets = total_dismissed.groupby('bowler').count().reset_index()
```

```python
a = count_of_wickets.sort_values(by = 'Wickets', ascending = False)
top_bowlers = a.loc[:,['bowler','Wickets']][0:10]
top_bowlers
```

#### üëâ So malinga is the most successfull bowler of all seasons of IPL taking a total of 170 wickets.
